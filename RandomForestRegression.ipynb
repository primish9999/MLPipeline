```python
import pandas as pd
import os
import biom
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Reading Metadata
mddat=pd.read_csv('/Users/priyanka99/Downloads/Metadata.tsv',sep='\t')
mddat
```

```python
mddat[mddat.columns[7]].isna().sum() # bmi has 87 null values
mddat.bmi.fillna(mddat.bmi.mean(), inplace=True)
#X and y , Train Test split
X = mddat[mddat.columns[669]]
y = mddat[mddat.columns[7]]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=400, test_size=200, random_state=4)
```

```python
#fitting

reg = RandomForestRegressor(random_state=1)
reg.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
```

```python
#Making Predictions

y = reg.predict(X_test)
```
